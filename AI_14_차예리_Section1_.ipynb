```python
import pandas as pd

df = pd.read_csv('vgames2.csv')
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
df.info()
```

```python
df[df.NA_Sales.str.contains('K')]
```

```python
df[df.NA_Sales.str.contains('M')]
```

```python
#전처리
df['Year'] = df['Year'].fillna(df['Year'].mean())  #평균으로 채워줌
df['Year'] = df['Year'].astype(int)

df['Genre'] = df['Genre'].fillna('X') 

freq = df.groupby('Publisher').size()
freq.sort_values(ascending=False)  
df['Publisher'] = df['Publisher'].fillna('electronic Arts')  #Publisher는 제일 많이 나온 electronic Arts로 채워줌
```

```python
df.info()
```

```python
def year(x):
  if x < 20 : 
    return  int(x + 2000)
  elif x >= 20 and x < 100: 
    return  int(x + 1900)
  else:
    return int(x)

df['Year'] = df['Year'].apply(year) #year에서 1,2자리의 숫자 발견되므로 4자리로 바꿔줌
```

```python
df['Year'].describe()
```

```python
#sales에서 k와 m을 없애고, sales의 object 데이터 형식을 float로 변환

sales = df.columns[-4:]

for i in sales:
  for j, v in enumerate(df[i]):
    if 'M' in df[i][j]:
      df[i][j] = float(v.replace('M',''))
    elif 'K' in df[i][j]:
      df[i][j] = float(v.replace('K',''))/1000
  df[i] = df[i].astype(float)
```

```python
#중복 제거
df = df.drop_duplicates(keep='last')
df.duplicated().sum()
```

```python
df_clean = df.copy()
df.info()
```

```python
#지역에 따라서 선호하는 게임 장르가 다를까

df_genre = df_clean[['Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
df_genre_group = df_genre.groupby(['Genre']).sum()
df_genre1 = df_genre_group.reset_index()
df_genre1
```

```python
#카이제곱 검정
#귀무가설 : 지역에 따라 선호하는 게임 장르는 똑같다.

from scipy.stats import chi2_contingency

chi, p, df, ex = chi2_contingency(df_genre_group)
print(p) #p-value는 0.05보다 작으므로 귀무가설 기각
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_genre1 = df_genre1.melt(id_vars = 'Genre', value_vars = ['NA_Sales', 'EU_Sales','JP_Sales','Other_Sales'])
df_genre1
```

```python
from matplotlib import RcParams
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(data=df_genre1, x="Genre", y = 'value',hue='variable')
```

```python
#연도별 게임의 트렌드가 있을까

df_year = pd.DataFrame(df_clean[['Year','Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']])
df_year['Sales_Sum'] = df_year['NA_Sales'] + df_year['EU_Sales'] + df_year['JP_Sales'] + df_year['Other_Sales']
df_year1 = df_year.groupby(['Year','Genre'])['Sales_Sum'].sum().reset_index()
```

```python
#귀무가설 : 연도에 따라 트렌드가 바뀌지 않는다.
import scipy.stats as stats

df_year2 = df_year1.groupby(['Year'])['Sales_Sum'].sum()
df_year2

print(stats.chisquare(df_year2))

#p-value는 0.05보다 작으므로 연도에 따라 트렌드가 바뀐다는 대립가설을 채택한다.
```

```python
#3. 출고량이 높은 게임에 대한 분석 및 시각화 프로세스

df_high = df_clean[['Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
df_high['Sales_Sum'] = df_high['NA_Sales']+df_high['EU_Sales']+df_high['JP_Sales']+df_high['Other_Sales']
df_high = df_high.sort_values(ascending=False, by=['Sales_Sum'])
df_high1 = df_high.head(10).reset_index().drop('index', axis=1)
df_high1 = df_high1[['Name', 'Sales_Sum']]
df_high1
```

```python
import numpy as np

cmap = plt.get_cmap("Set3")
color = cmap(np.array([3,2,0,4,5,6,7,8,1,10]))
plt.pie(df_high1['Sales_Sum'], labels = df_high1['Name'], 
        autopct='%.1f%%', 
        explode=(0.1,0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02),
        colors=color)
plt.title('Top 10 Game Sales', fontsize=25)
plt.show()
```

```python
df_recent = df_clean.query("Year >= 2010").reset_index().drop('index', axis=1)
df_recent
```

```python
#최근 10년 지역별 선호 장르
df_Genre = df_recent[['Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
df_Genre_group = df_Genre.groupby('Genre')['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'].sum()
df_Genre1 = df_Genre_group.reset_index()

df_Genre1.plot.barh('Genre',figsize=(15,13))
```

```python
#최근 10년 지역별 선호 플랫폼
df_platform = df_recent[['Platform', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
df_platform_NA = df_platform[['Platform', 'NA_Sales']]
df_platform_EU = df_platform[['Platform', 'EU_Sales']]
df_platform_JP = df_platform[['Platform', 'JP_Sales']]
df_platform_Other = df_platform[['Platform', 'Other_Sales']]

df_platform_NA = df_platform_NA.groupby('Platform').sum().sort_values(by='NA_Sales', ascending=False)
df_platform_NA = df_platform_NA.head(5).reset_index()

df_platform_EU = df_platform_EU.groupby('Platform').sum().sort_values(by='EU_Sales', ascending=False)
df_platform_EU = df_platform_EU.head(5).reset_index()

df_platform_JP = df_platform_JP.groupby('Platform').sum().sort_values(by='JP_Sales', ascending=False)
df_platform_JP = df_platform_JP.head(5).reset_index()

df_platform_Other = df_platform_Other.groupby('Platform').sum().sort_values(by='Other_Sales', ascending=False)
df_platform_Other = df_platform_Other.head(5).reset_index()

f,axes = plt.subplots(2,2)
f.set_size_inches((15,10))

axes[0,0].bar(df_platform_NA['Platform'], df_platform_NA['NA_Sales'],color='#F78181')
axes[0,0].set_title('NA_Sales')

axes[0,1].bar(df_platform_EU['Platform'], df_platform_EU['EU_Sales'], color='#F7BE81')
axes[0,1].set_title('EU_Sales')

axes[1,0].bar(df_platform_JP['Platform'], df_platform_JP['JP_Sales'], color='#F5DA81')
axes[1,0].set_title('JP_Sales')

axes[1,1].bar(df_platform_Other['Platform'], df_platform_Other['Other_Sales'], color='#819FF7')
axes[1,1].set_title('Other_Sales')

plt.show()
```
